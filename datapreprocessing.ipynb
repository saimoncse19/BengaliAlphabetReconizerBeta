```python
project_path = 'BanglaLekha-Isolated/Images'
```

```python
import numpy as np
import cv2
import os
import glob
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
categories = [i for i in range(1, 82)]
categories
```

```python
img_size = 30

for category in categories:
  path = os.path.join(project_path, str(category))

  for img in os.listdir(path):
    img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
    img_arr = cv2.resize(img_arr, (img_size, img_size))
    print(img_arr.shape)
    plt.imshow(img_arr, cmap='gray')
    plt.show()
    break
  break
```

```python
data = []

def get_data():
  for category in categories:
    path = os.path.join(project_path, str(category))

    for img in os.listdir(path):
      img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
      img_arr = cv2.resize(img_arr, (img_size, img_size))
      data.append([img_arr, category])


get_data()
```

```python
import pickle
pickle.dump(data,open("dataUpdated30.pickle","wb"))
```
